```python
def convert_units(x):
    try:
        return float(x.replace("MB", "000000").replace("KB", "000").replace("B", ""))/1000000
    except:
        return -1

    
DIR = "/home/sala/Work/Data/Network/"
# Some basic and horrible data wrangling is needed before using pandas
# grep -v "#" iftop_20150206_1007_LCLStoPSI.log | grep -v "\---" | grep -v "rate" | grep -v "==" | grep -v "Cumulative" \n
#| grep -v '^$' | sed "s=^   ==g" | sed "s=  *=:=g" > iftop_20150206_1007_LCLStoPSI_norm.log 

# loading normalized iftop output
fname = "/home/sala/Work/Data/Network/iftop_20150206_0930_LCLStoPSI_norm.log"
df = pd.read_csv(fname, sep=':', skiprows=1, names=["id", "host", "dir", "2s", "10s", "40s", "cum"], 
                 converters={"2s": convert_units, "10s":convert_units, "40s":convert_units}, error_bad_lines=False)
df.ix[1::2,"id"].values[:] = df[0::2]["id"].values[:]

fname = "/home/sala/Work/Data/Network/iftop_20150206_1007_LCLStoPSI_norm.log"
df2 = pd.read_csv(fname, sep=':', skiprows=1, names=["id", "host", "dir", "2s", "10s", "40s", "cum"], 
                  converters={"2s": convert_units, "10s":convert_units, "40s":convert_units}, error_bad_lines=False)
df2.ix[1::2,"id"].values[:] = df2[0::2]["id"].values[:]

df = df.append(df2)

# reshuffle destination and host columns
dest = df["host"].values[:].copy()
dest[0::2] = dest[1::2]
df['dest'] = pd.Series(dest, index=df.index)
df["host"][1::2].values[:] = df["host"][0::2].values[:]
cols = df.columns.tolist()
cols = cols[:3] + [cols[-1],] + cols[3:-2]
df = df[cols]

print df[0:10]
```

```python
data = df[(df["dest"]=="psexport02.slac.stanford.edu") & (df["dir"] == "<=")]
plt.plot(range(0, data["2s"].shape[0]*2, 2), data["2s"], label="psexport02 (LCLS) -> xbl-export-1")
plt.title("GlobusOnline traffic (iftop)")
plt.xlabel("seconds")
plt.ylabel("MB/s")
plt.legend(loc="best")
plt.show()

data2 = df[ (df["dest"]=="mx-fs-1.psi.ch") & (df["dir"]=="=>")]
plt.plot(range(0, data2["2s"].shape[0]*2, 2), data2["2s"], label="xbl-export-1 -> fs")
data3 = df[ (df["dest"]=="mx-fs-1.psi.ch") & (df["dir"]=="<=")]
plt.plot(range(0, data3["2s"].shape[0]*2, 2), data3["2s"], label="xbl-export-1 <- fs")
plt.title("NFS traffic (iftop)")
plt.xlabel("seconds")
plt.ylabel("MB/s")
plt.legend(loc="best")
plt.show()
```
